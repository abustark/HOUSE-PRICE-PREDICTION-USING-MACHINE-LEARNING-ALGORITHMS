```python
#======================= IMPORT PACKAGES ============================
import pandas as pd
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

#======================= DATA SELECTION =========================

print("=======================================")
print("---------- Data Selection -------------")
print("=======================================")
data=pd.read_csv('Bengaluru_House_Data.csv')
print(data.head(10))
print()
```

```python
#==================== PREPROCESSING =======================================

#checking missing values

print("=====================================================")
print("--------- Before Checking missing values ------------")
print("=====================================================")
print(data.isnull().sum())
print()


print("=====================================================")
print("--------- After Checking missing values ------------")
print("=====================================================")
data=data.fillna(0)
print(data.isnull().sum())
print()
```

```python
#==== MODIFIED LABEL ENCODING (Create and Fit Separate Encoders) ====

from sklearn import preprocessing
import joblib # Make sure joblib is imported

# Create a dictionary to store encoders
label_encoders = {}
categorical_cols = ['area_type', 'availability', 'location', 'size', 'society']

print("Fitting and saving Label Encoders...")
for col in categorical_cols:
    print(f"Processing column: {col}")
    # Create a new encoder for each column
    le = preprocessing.LabelEncoder()
    # Fit the encoder and transform the column in the dataframe
    # Convert column to string first to handle potential mixed types or NaNs safely
    data[col] = le.fit_transform(data[col].astype(str))
    # Store the fitted encoder
    label_encoders[col] = le
    # Save the fitted encoder
    joblib.dump(le, f'le_{col}.joblib') # Save with a specific name, e.g., le_area_type.joblib

print("--- Label Encoders fitted and saved ---")
print("Sample of encoded 'area_type':")
print(data['area_type'].head(5)) # Verify encoding still works

# --- Handle total_sqft (Keep your original logic if it worked) ---
# Be careful here - dropping it means the model won't use it.
# If you intended to convert it to numeric, that needs to happen *before* splitting data.
# Example conversion (if it's mostly numbers but some ranges like '1000 - 1200'):
# def convert_sqft_to_num(x):
#     tokens = x.split('-')
#     if len(tokens) == 2:
#         return (float(tokens[0]) + float(tokens[1])) / 2
#     try:
#         return float(x)
#     except:
#         return None # Handle non-numeric cases
# data['total_sqft'] = data['total_sqft'].apply(convert_sqft_to_num)
# data = data.dropna(subset=['total_sqft']) # Drop rows where conversion failed

# If you are dropping it as before:
try:
    data = data.drop('total_sqft', axis=1)
    print("Dropped 'total_sqft' column.")
except KeyError:
    print("'total_sqft' column not found or already dropped.")
```

```python
#========================= DATA SPLITTING ============================

#=== TEST AND TRAIN ===

x=data.drop('price',axis=1)
y=data['price']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)

print("-----------------------------------------------------------")
print("======================= Data splitting ====================")
print("-----------------------------------------------------------")
print()
print("Total No Of data          :",data.shape[0])
print()
print("Total No of Training data :",X_train.shape[0])
print()
print("Total No of Testing data :",X_test.shape[0])
print()
```

```python
#========================= CLASSIFICATION ============================

from sklearn.linear_model import Ridge
from sklearn import metrics

#=== ridge regression ===

#initialize the model
ridgeR = Ridge(alpha = 1)

#fitting the model
ridgeR.fit(X_train, y_train)

#predict the model
y_pred = ridgeR.predict(X_test)


print("-----------------------------------------------------------")
print("======================= RIDGE REGRESSION ===================")
print("-----------------------------------------------------------")
print()


mae_ridge=metrics.mean_absolute_error(y_test, y_pred)

print("1.Mean Absolute Error : ",mae_ridge)

```

```python
#===== Random Forest Regression Model =======

from sklearn.ensemble import RandomForestRegressor
 
 # create regressor object
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
 
# fit the regressor with x and y data
regressor.fit(X_train, y_train) 

# --- ADD THIS LINE TO SAVE THE MODEL ---
joblib.dump(regressor, 'model.joblib')
print("Random Forest model saved as model.joblib")
# ----------------------------------------
y_pred_rf=regressor.predict(X_test)

print("-----------------------------------------------------------")
print("============== RANDOM FOREST REGRESSION ===================")
print("-----------------------------------------------------------")
print()


mae_rf=metrics.mean_absolute_error(y_pred_rf,y_test)

print("1.Mean Absolute Error :",mae_rf)
print()
```

```python
#========================= PREDICTION ============================

print("-----------------------------------------------------------")
print("======================= PREDICTION ========================")
print("-----------------------------------------------------------")
print()

for i in range(0,10):
    Results=y_pred_rf[i]
    print("------------------------------------------")
    print()
    print([i],"The predicted house price is ", Results)
    print()
```
